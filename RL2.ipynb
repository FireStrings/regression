```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataset = pd.read_csv('data/houses_to_rent.csv', encoding="latin1")
dataset.rename(columns={"city": "cidade", "rooms": "quartos", "bathroom": "banheiros", 
"parking spaces": "vagas est.", "floor":"andar", "furniture": "mobilhado",
"hoa":"valor cond.", "rent amount":"aluguel", "property tax":"iptu", "fire insurance":"seguro incendio", }, inplace=True)

dataset.head()
```

```python


# for i in dataset.columns:
# #     if dataset[i].dtype == np.int64:
        
# #         for j in dataset.columns:
# #             if dataset[j].dtype == np.int64:
# #                 print(i)
#     print(i)
#     print(dataset[i].dtype)

def _fmt(s):
    if s == "Incluso":
        return 0
    if s == "-":
        return 0
    
    
    return float(s.split("R$")[1].replace(",", "."))

dataset["aluguel"] = dataset["aluguel"].apply(_fmt)

X = dataset["andar"].values
Y = dataset["aluguel"].values


# for i in range(0, 50):
#     print(X[i], Y[i])
print(np.corrcoef(X,Y))

# print(np.log())
# print("\n")


#plotting line 
# plt.plot(x, y, color='#00ff00', label='Linear Regression')#plot the data point
plt.scatter(np.log(X), np.log(Y), color='#ff0000', label='Data Point')# x-axis label
# plt.scatter(X, Y, color='#ff0000', label='Data Point')# x-axis label
plt.xlabel('Aluguel')#y-axis labelevenly
plt.ylabel('Area')
plt.legend()
plt.show()

```
