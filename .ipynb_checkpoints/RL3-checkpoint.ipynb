```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt

base = pd.read_csv("data/plano_saude.csv")

X = base.iloc[:,0].values
y = base.iloc[:,1].values

correlacao = np.corrcoef(X,y)

print("Correlação Linear: " + str(correlacao[0][1]))

X = X.reshape(-1,1)

regressor = LinearRegression()

regressor.fit(X,y)

# print(regressor.coef_)
# print(regressor.intercept_)

plt.scatter(X,y)
plt.plot(X, regressor.predict(X), color="red")
plt.title("Regressão simples")
plt.xlabel("Idade")
plt.ylabel("Custo")
anos = 90
previsao1 = regressor.predict([[anos]])
# previsao2 = regressor.intercept_ + regressor.coef_ * 40
print("Previsão de valor para " + str(anos) + " anos: " + str(previsao1[0]))
# print(previsao2)

score = regressor.score(X, y)

print(score)
```
